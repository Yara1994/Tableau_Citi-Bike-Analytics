```python
# Import Dependencies

import pandas as pd
```

```python
# Import CSV files and save variables

january_2018 = pd.read_csv("Resources/Raw_Data/JC-201801-citibike-tripdata.csv")
february_2018 = pd.read_csv("Resources/Raw_Data/JC-201802-citibike-tripdata.csv")
march_2018 = pd.read_csv("Resources/Raw_Data/JC-201803-citibike-tripdata.csv")
april_2018 = pd.read_csv("Resources/Raw_Data/JC-201804-citibike-tripdata.csv")
may_2018 = pd.read_csv("Resources/Raw_Data/JC-201805-citibike-tripdata.csv")
june_2018 = pd.read_csv("Resources/Raw_Data/JC-201806-citibike-tripdata.csv")
july_2018 = pd.read_csv("Resources/Raw_Data/JC-201807-citibike-tripdata.csv")
august_2018 = pd.read_csv("Resources/Raw_Data/JC-201808-citibike-tripdata.csv")
september_2018 = pd.read_csv("Resources/Raw_Data/JC-201809-citibike-tripdata.csv")
october_2018 = pd.read_csv("Resources/Raw_Data/JC-201810-citibike-tripdata.csv")
november_2018 = pd.read_csv("Resources/Raw_Data/JC-201811-citibike-tripdata.csv")
december_2018 = pd.read_csv("Resources/Raw_Data/JC-201812-citibike-tripdata.csv")
```

```python
# Combine all datasets in one

full_year = pd.concat([january_2018, february_2018, march_2018, april_2018, may_2018, june_2018, july_2018, august_2018, september_2018, october_2018, november_2018, december_2018], join = "outer")
full_year
```

```python
# Create another column with Trip Duration in minutes for better understanding

tripduration_minutes = []

for tripduration in full_year["tripduration"]:
    tripduration_minutes.append(round(tripduration / 60, 1))
    
full_year["Trip Length in minutes"] = tripduration_minutes

    
```

```python
full_year
```

```python
# Create another column with Genres for better understanding

genders = []

for gender in full_year["gender"]:
    
    if gender == 1:
        genders.append("Male")
        
    elif gender == 2:
        genders.append("Female")
    else:
        genders.append("Unknown")

full_year["Gender"] = genders
```

```python
full_year
```

```python
# Pick only columns we need

updated_full_year_2018 = full_year[["tripduration", "Trip Length in minutes", "starttime", "stoptime", "start station id", "start station name", "start station latitude", "start station longitude", "end station id", "end station name", "end station latitude", "end station longitude", "bikeid", "usertype", "birth year", "Gender"]]
updated_full_year_2018
```

```python
# Rename all columns

final_update = updated_full_year_2018.rename(columns = {"tripduration" : "Trip Length in seconds",
                                        "starttime": "Start Time",
                                        "stoptime" : "Stop Time",
                                        "start station id" : "Start Station ID",
                                        "start station name": "Start Station Name",
                                        "start station latitude": "Start Station Latitude",
                                        "start station longitude": "Start Station Longitude",
                                        "end station id": "End Station ID",
                                        "end station name": "End Station Name",
                                        "end station latitude": "End Station Latitude",
                                        "end station longitude": "End Station Longitude",
                                        "bikeid": "Bike ID",
                                        "usertype": "User Type",
                                        "birth year": "Birth Year" 
                                       })
```

```python
final_update
```

```python
# Save updated dataframe in CSV

final_update.to_csv("Resources/2018_full_trip_information.csv", index = False)
```
